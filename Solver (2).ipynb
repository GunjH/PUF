```python
def Solver(X,Y):
  
  eta = 0.1              #learning rate
  epochs = 3500          #no of iterations
  lambda_para = 0.001              #hyper parmeter

  #train svm

  (n,d) = X.shape         #rows=10000, coukmns=729 
  #print('d',d)

  #initialize weight
  #W = np.random.normal(loc=0, scale=1, size=d)            #loc = mean = 0; scale = std = 1
  W = np.zeros(d)                                  #b appended to W vector; b=0


  #stochastic gradient descent for opt

  for itr in range (1,epochs):
    i=random.randint(0,Y.shape[0]-1)

    discriminent= Y[i] * np.dot(X[i],W)

    if (discriminent >=1):
      dw = lambda_para * W    
    else:
      dw = lambda_para * W - Y[i] * X[i] 

    W = W - eta/(np.sqrt(itr)) * dw
    #W = W - eta * dw

  return W

```

```python
feat_X=get_features(X)

weight=Solver(feat_X,Y)

#weight
```
